# Convert date to datetime64
We can see from the code below that the date ('DATETIMEINIT' and 'DATETIMECLOSED' are actually identified as a str, not dt

```python
oakland_311_calls_original['DATETIMEINIT'].min ()
```

To correct this, we have to first create a new dataframe from the original so that we always have original data then use pd.to_datetime to convert for both 'DATETIMEINIT' and 'DATETIMECLOSED'

```python
oakland_311 = oakland_311_calls_original.copy()
```

```python
oakland_311['DATETIMEINIT'] = pd.to_datetime(oakland_311['DATETIMEINIT'])
```

```python
oakland_311['DATETIMECLOSED'] = pd.to_datetime(oakland_311['DATETIMECLOSED'])
```

```python
oakland_311.head()
```

```python
oakland_311.tail()
```

Now to confirm whether the dtype for dates changed in the dataframe

```python
oakland_311.info()
```

# Finding the earliest  and latest dates
We will use .min and .max to establish the oldest call and the newest.

```python
oakland_311['DATETIMEINIT'].min()
```

```python
oakland_311['DATETIMECLOSED'].max()
```

From this, we can establish that the first call in the data was made in July, 2009 while the latest was made on Oct 29, 2021.

# Difference between the two dates
To find out how much time has lapsed between the two, we will create a new column in the date called 'DATECLOSED' and populate it with the difference between 'DATETIMECLOSED' and 'DATETIMEINIT'

```python
oakland_311['DATECLOSED'] = oakland_311['DATETIMECLOSED'] - oakland_311['DATETIMEINIT']
```
